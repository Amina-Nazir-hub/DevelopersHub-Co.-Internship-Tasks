```python
"""
Introduction and Problem Statement
The goal is to understand how to read, summarize, and visualize a dataset. I am using the Iris dataset, which contains measurements of different 
flower species.
"""
```

```python
"""
Dataset Understanding
The Iris dataset includes:
Sepal Length
Sepal Width
Petal Length
Petal Width
Species (Setosa, Versicolor, Virginica)
"""
```

```python
#Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
#Load Dataset
df = sns.load_dataset("iris")
```

```python
#Dataset Structure 
# Shape of dataset
print("Dataset Shape:", df.shape)
# Column names
print("Columns:", df.columns)
# Dataset info
df.info()
# Summary statistics
df.describe()
```

```python
#Data Cleaning & Preparation
# Check for missing values
print(df.isnull().sum())
# Check for duplicates
print("Duplicate rows:", df.duplicated().sum())
```

```python
"""
Exploratory Data Analysis (EDA)
Scatter Plot
Relationship between petal length and petal width.
"""

plt.figure(figsize=(6,4))
sns.scatterplot(x="petal_length", y="petal_width", hue="species", data=df)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
"""
Histogram
Distribution of sepal length.
"""

plt.figure(figsize=(6,4))
sns.histplot(df["sepal_length"], kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
"""
Box Plot
Detecting outliers and spread of sepal width.
"""

plt.figure(figsize=(6,4))
sns.boxplot(x="species", y="sepal_width", data=df)
plt.title("Sepal Width Spread Across Species")
plt.show()
```

```python
"""
Conclusion
The Iris dataset is clean and easy to explore.
Scatter plots help understand relationships between features.
Histograms show the distribution of individual features.
Box plots identify outliers and feature spread.
Basic EDA provides insights about the data before any further analysis.
"""
```
